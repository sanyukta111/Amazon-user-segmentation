```python
import pandas as pd

orders = pd.read_csv('orders.csv')
products = pd.read_csv('products.csv')
```

```python
# Performing a Left Join to merge orders with product details
df = pd.merge(orders, products, on='product_id', how='left')

# Displaying the result
print("Best Join (Left Join) for Our Use Case:")
df.head()
```

# Handling missing values

```python
df.info()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
mean_rating = df['rating'].mean()
df['rating'] = df['rating'].fillna(mean_rating)
```

```python
df.isnull().sum()
```

```python
df['category'].fillna('Unknown', inplace=True)
```

```python
df.info()
```

# Dates

```python
df['order_timestamp']
```

```python
df['order_timestamp'] = pd.to_datetime(df['order_timestamp'])
df.info()
```

```python
df['order_timestamp'].dt.year
```

```python
df['order_timestamp'].dt.month
```

```python
df['order_timestamp'].dt.day_name()
```

```python
#Example 1: Extract week number
df['week_number'] = df['order_timestamp'].dt.isocalendar().week

# Example 2: Extract day of the week
df['day_name'] = df['order_timestamp'].dt.day_name()

# Example 3: Extract year and month
df['year'] = df['order_timestamp'].dt.year
df['month'] = df['order_timestamp'].dt.month

# Display results
print("Data with DateTime Components Extracted:")
print(df[['order_timestamp', 'week_number', 'day_name', 'year', 'month']].head())
```

```python
df.info()
```

```python
# get a column for month-year
df['month_year'] = df['order_timestamp'].dt.strftime('%b-%y')
df['month_year'].head()
```

# Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x='rating', bins=15, kde=True)
```

```python
mean_rating = df['rating'].mean()
plt.figure(figsize=(8, 6))
plt.hist(df['rating'], bins=10, color='skyblue')
plt.axvline(mean_rating, color='red', linestyle='--', label=f"Mean: {mean_rating:.2f}")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# analyse the distribution of category variable
sns.countplot(data=df, y='category')
```

```python
sns.countplot(data=df, x='category', palette='coolwarm')
plt.xticks(rotation=45)
plt.show()
```

```python
categories = df['category'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(categories.index, categories.values, color='skyblue')
plt.xticks(rotation=45)
plt.title("Count of Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# numeric vs categorical
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='category', y='rating', palette='Set3') # Styling Tip 💡 palette

plt.xticks(rotation=45)
plt.title("Rating Distribution by Category", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Rating", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='rating_category', y='discounted_price', inner='quartile', palette='Accent')

plt.title("Violin Plot of Discounted Price by Rating Category", fontsize=16)
plt.xlabel("Rating Category", fontsize=12)
plt.ylabel("Discounted Price", fontsize=12)
plt.show()
```

```python
# how ratings vary across category for durable and non-durable items
# 1. rating
# 2. category
# 3. durability
# numeric vs categorical vs categorical

grouped_data = df.groupby(['category','contains_durable'])['rating_filled'].mean().reset_index()
```
